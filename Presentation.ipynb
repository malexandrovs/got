```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ptitprince as pt
import seaborn as sns
sns.set()
from PIL import Image
from scipy import stats
from statsmodels.stats.anova import AnovaRM
from statsmodels.stats.api import anova_lm
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
battles = pd.read_csv('battles.csv')
battles.head()
```

```python
deaths = pd.read_csv('character-deaths.csv')
deaths.head(10)
```

```python
pred = pd.read_csv('character-predictions.csv')
pred.head()
```

```python
model = ols('popularity ~ isAlive',pred)
res = model.fit()
table = sm.stats.anova_lm(res, typ=1)
sns.violinplot(pred['isAlive'], pred['popularity'])
table
```

```python
newFirst = pred['name'].str.split(" ", n = 1, expand = True)
pred['First Names'] = newFirst[0]
newLast = pred['name'].str.rsplit(" ", 1, expand = True)
pred['Last Names'] = newLast[1]
```

```python
predGrouped = pred.groupby(['First Names'])
predGrouped = predGrouped.filter(lambda x: x['First Names'].count() > 6)

plt.figure(figsize=(16, 6))
sns.set(style="darkgrid")
plt.yticks(np.arange(1, 22, step=2))

sns.countplot(x="First Names", data=predGrouped, order = predGrouped['First Names'].value_counts().index)
plt.ylabel('Number of Characters with ths Name')
plt.show()
```

```python
new = deaths['Allegiances'].str.split(" ", n = 1, expand = True)
new = new.replace({'House': None, "Night's": "Night's Watch" })
new[0] = new[0].fillna(new[1])
deaths['Allegiances Clean'] = new[0]

plt.figure(figsize=(25, 6))
sns.set(style="darkgrid")
plt.xlabel('House')
sns.barplot(x=deaths["Allegiances Clean"].value_counts().index, y=deaths['Allegiances Clean'].value_counts()/deaths['Name'].count()*100, order = deaths['Allegiances Clean'].value_counts().index)
plt.ylabel('Deaths in %')
plt.show()
```

```python

fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(25, 20), sharex=True)
ax = sns.lineplot(x=deaths['Death Year'].value_counts().index, y=deaths['Death Year'].value_counts(), ax=axes[0])
ax.yaxis.tick_right()
ax.set_ylabel('Death Count')
ax = sns.scatterplot(x=battles['year'], y=battles['name'], ax=axes[1])

ax.set_xticks(np.arange(297, 301, step=1))
ax.yaxis.tick_right()
ax.set_ylabel('Battles')
ax.set_xlabel('Year')
fig.tight_layout()
```
